```python
#!pip install jieba
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from wordcloud import WordCloud, STOPWORDS
import jieba
import re
```

```python
import requests
from bs4 import BeautifulSoup
url = "https://www.sohu.com/a/226127592_139908"
responce = requests.get(url)
soup = BeautifulSoup(responce.text,'lxml')    #lxml用来解析网页
responce.status_code #如果返回值为200则表示访问成功
```

```python
links_div = soup.find("article",class_="article")
#links_div = soup.find(class_="content")
print(type(links_div))     #我们可以尝试输出下获取的文本，验证是否是我们想要的
links_div=str(links_div)
text = links_div.strip()   #对字符串进行切片操作
#text=links_div
#text
```

```python
dic = {text}
with open(r'dataset\report.txt','a',encoding='utf-8')as f:
    f.write(str(dic))
f.close()
```

```python

r ='[，。\%、；1234567890n]'
file=open(r'dataset\report.txt','r',encoding='utf-8').read()
file =re.sub(r,'',file)      #剔除无关信息
con = jieba.lcut(file)     #分词
words = " ".join(con)    #分词后插入空格
words
```

```python
wordcloud = WordCloud(font_path="simkai.ttf",background_color="white",width=800, height=660).generate(words) 
wordcloud.to_file(r'dataset\2017_gov_report.png')     #保存图片
```

```python
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

# display the cloud
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
alice_mask = np.array(Image.open(r'.\dataset\alice_mask.png'))
alice_wc = WordCloud(background_color='white', max_words=2000, mask=alice_mask,font_path="simkai.ttf")

# generate the word cloud
alice_wc.generate(words)

# display the word cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
