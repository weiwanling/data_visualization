```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_can=pd.read_excel('./dataset/Canada.xlsx',
                     sheet_name='Canada by Citizenship',
                     skiprows=20,
                     skipfooter=2
                    )
```

```python
years=list(range(1980,2014))
df_tot=pd.DataFrame(df_can[years].sum(axis=0))
df_tot.reset_index(inplace=True)
df_tot.index=map(int,df_tot.index)
df_tot.columns=['year','total']
df_tot.head()
```

```python
plt.figure(figsize=(15,10))
sns.set_style('whitegrid')   #darkgrid  whitegrid ticks
sns.set(font_scale=1.5)   # set size x,y label title
ax=sns.regplot(data=df_tot,x='year',y='total',color='green',marker='+',scatter_kws={'s':20}) # set marker size
ax.set(xlabel='Year', ylabel='Total Immigration')
ax.set_title('Total Immigrationn from all countries to Canada from 1980 - 2013')
```
