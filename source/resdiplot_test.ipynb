<h3>resdiplot使用 </h3>
<p>This function will regress y on x (possibly as a robust or polynomial regression) and 
then draw a scatterplot of the residuals. You can optionally fit a lowess smoother to the residual plot
, which can help in determining if there is structure to the residuals.
</p>

```python
%matplotlib inline
import numpy as np
import seaborn as sns
sns.set(style="whitegrid")

# Make an example dataset with y ~ x
rs = np.random.RandomState(7)
x = rs.normal(2, 1, 75)
y = 2 + 1.5 * x + rs.normal(0, 2, 75)

# Plot the residuals after fitting a linear model
sns.residplot(x, y, lowess=True, color="g")
```
